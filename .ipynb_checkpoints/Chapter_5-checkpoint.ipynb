```python
%matplotlib inline
import pymc3 as pm
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

# %config InlineBackend.figure_format = 'retina'
# plt.style.use(['seaborn-colorblind', 'seaborn-darkgrid'])
```

```python
#R code 5.1
d = pd.read_csv('Data/WaffleDivorce.csv', sep=';', header=0)
d['age_norm'] = (d.MedianAgeMarriage - d.MedianAgeMarriage.mean())/d.MedianAgeMarriage.std()
d['divorce_norm'] = (d.Divorce - d.Divorce.mean())/d.Divorce.std()
d['marriage_norm'] = (d.Marriage - d.Marriage.mean())/d.Marriage.std()
```

```python
#R code 5.2
d.MedianAgeMarriage.std()
```

```python
#R code 5.3 (out of order intentionally)
with pm.Model() as m_5_1:
    alpha = pm.Normal('alpha', mu=0, sd=0.2) 
    beta_age = pm.Normal('beta_age', mu=0, sd=0.5) 
    mu = pm.Deterministic('mu', alpha + beta_age*d.age_norm)
    sigma = pm.Exponential('sigma', 1)
    divorce = pm.Normal('divorce', mu=mu, sd=sigma, observed=d.divorce_norm)
    trace_5_1 = pm.sample(1000, tune=1000, chains=1) 
    prior = pm.sample_prior_predictive()
```

```python
varnames = ['alpha','beta_age','sigma']
pm.summary(trace_5_1, varnames, alpha = 0.11).round(2)
```

```python
#R code 5.4. There are probably better ways to do this.
x = np.linspace(-2, 2, 100)
z=0
while z < 50:
    plt.plot(x, prior['alpha'][z] + prior['beta_age'][z]*x, linewidth=0.1, color='b')
    z+=1
plt.xlabel('Marriage Age Normalized')
plt.ylabel('Divorce Rate Normalized');

#The graph below shows that the priors we chose were bad. The standard deviation of the median age of marriage (not
#normalized) is 1.2 (R code 5.2). So, when the normalized marriage age is 1.0 (corresponding to an increase of age)
#of 1.2 years, our priors permit an increase in divorce rate by one full standard deviation.
```

```python
#R code 5.5, figure 5.2 (right)

mu_mean = trace_5_1['mu']
mu_hpd = pm.hpd(mu_mean)

plt.plot(d.age_norm, d.divorce_norm, 'C0o')
plt.plot(d.age_norm, mu_mean.mean(0), 'C2')

idx = np.argsort(d.age_norm)
plt.fill_between(d.age_norm[idx],
                 mu_hpd[:,0][idx], mu_hpd[:,1][idx], color='C2', alpha=0.25)

plt.xlabel('Meadian Age Marriage', fontsize=14)
plt.ylabel('Divorce', fontsize=14);
```

```python
#5M4
d = pd.read_csv('Data/WaffleDivorce.csv', sep=';', header=0)
d['age_norm'] = (d.MedianAgeMarriage - d.MedianAgeMarriage.mean())/d.MedianAgeMarriage.std()
d['divorce_norm'] = (d.Divorce - d.Divorce.mean())/d.Divorce.std()
d['marriage_norm'] = (d.Marriage - d.Marriage.mean())/d.Marriage.std()
d2 = pd.read_csv('Data/Mormons.csv', sep=',', header=0)

d3 = pd.merge(left=d, right=d2, how='left', left_on='Location', 
              right_on='State').drop(['State'], axis=1)

d3['M_percent'] = [i.strip('%') for i in d3['M_percent']]
d3['M_percent'] = [float(i) for i in d3['M_percent']]

d3['mormon_norm']=(np.log(d3.M_percent) - np.log(d3.M_percent).mean())/np.log(d3.M_percent).std()

with pm.Model() as m_5M4:
    alpha = pm.Normal('alpha', mu=0, sd=0.2) 
    beta_marriage = pm.Normal('beta_marriage', mu=0, sd=0.5) 
    beta_age = pm.Normal('beta_age', mu=0, sd=0.5)
    beta_mormon = pm.Normal('beta_mormon', mu=0, sd=0.5)
    mu = pm.Deterministic('mu', alpha + beta_marriage*d3.marriage_norm + 
                          beta_age*d3.age_norm + beta_mormon*d3.mormon_norm)
    sigma = pm.Exponential('sigma', 1)
    divorce = pm.Normal('divorce', mu=mu, sd=sigma, observed=d3.divorce_norm)
    trace_5M4 = pm.sample(1000, tune=1000, chains=1) 
```

```python
mu_mean = trace_5M4['mu']
mu_hpd = pm.hpd(mu_mean)
x = np.linspace(-1,4)
plt.plot(d3.mormon_norm, d3.divorce_norm, 'C0o')
plt.plot(x, trace_5M4['alpha'].mean() + trace_5M4['beta_mormon'].mean()*x)
plt.plot(x, trace_5M4['alpha'].mean() + trace_5M4['beta_marriage'].mean()*x)
#plt.plot(d3.mormon_norm, mu_mean.mean(0), 'C2')

# idx = np.argsort(d3.mormon_norm)
# plt.fill_between(x, mu_hpd[:,0][idx], mu_hpd[:,1][idx], color='C2', alpha=0.25)

plt.xlabel('Mormon pct, standardized', fontsize=14)
plt.ylabel('Divorce', fontsize=14);
```

```python
#Now the correlations seem meaningful. Maybe a forest plot will help:
varnames = ['beta_area', 'beta_size']
pm.forestplot((trace_5H1_1, trace_5H1_2, trace_5H2), model_names = ("Only area","Only size","Both"), 
              var_names=varnames, credible_interval=0.89, linewidth=1, figsize=(6,2));
#I think the punch line is: knowing just territory area or group size isn't helpful, but once you can control
#for one of them, the other one becomes a good predictor of weight. More specifically, if all foxes have an
#average pack size, then those with a larger area to hunt will be heavier, more likely because there's more food
#and you're sharing it over the average pack size. On the other hand, if you control for area, which means 
#giving all packs the same hunting area, then the larger the pack size, the smaller the fox weight, because
#the food will have to be shared across more foxes.
```

```python
#5H3
#First the model that looks at body weight as a function of avg food and group size
with pm.Model() as m_5H3_1:
    alpha = pm.Normal('alpha', mu=0, sd=10) 
    beta_food = pm.Normal('beta_food', mu=0, sd=4)
    beta_size = pm.Normal('beta_size', mu=0, sd=4)
    mu = pm.Deterministic('mu', alpha + beta_food*d.avgfood_s + beta_size*d.groupsize_s)
    sigma = pm.Exponential('sigma', 1)
    weight = pm.Normal('weight', mu=mu, sd=sigma, observed=d.weight_s)
    trace_5H3_1 = pm.sample(1000, tune=1000, chains=1) 
```

```python
#Now looking at body weight as a function of avg food, group size, and area
with pm.Model() as m_5H3_2:
    alpha = pm.Normal('alpha', mu=0, sd=10) 
    beta_food = pm.Normal('beta_food', mu=0, sd=4)
    beta_size = pm.Normal('beta_size', mu=0, sd=4)
    beta_area = pm.Normal('beta_area', mu=0, sd=4)
    mu = pm.Deterministic('mu', alpha + beta_food*d.avgfood_s + beta_size*d.groupsize_s
                         + beta_area*d.area_s)
    sigma = pm.Exponential('sigma', 1)
    weight = pm.Normal('weight', mu=mu, sd=sigma, observed=d.weight_s)
    trace_5H3_2 = pm.sample(1000, tune=1000, chains=1) 
```

```python
#A forest plot to help visualize
varnames = ['beta_area', 'beta_size', 'beta_food']
pm.forestplot((trace_5H1_1, trace_5H1_2, trace_5H2, trace_5H3_1, trace_5H3_2), 
              model_names = ("Only area","Only pack size","Area and pack size",
                             "Food and pack size","Food, pack size, and area"), 
              var_names=varnames, credible_interval=0.89, linewidth=1, figsize=(6,5));
#If I had to pick a model, I'd use the one that just looks at average food and pack size. There's overlap in the
#variables. For example, average food is probably a function of pack size and area. I'm sure a counterfactual
#plot would support this idea.
```
