```python
def precis(data, show_hist=False):
    if show_hist:
        plt.hist(data, int(np.sqrt(len(data)).round()))
    return f'Mean: {np.mean(data)}, std: {np.std(data)}, 5.5%: {np.percentile(data, 5.5)}, 94.5%: {np.percentile(data, 94.5)}'
```

```python
#just an exercise to basically replicate the binomial distribution, I think.
def walk_int_steps(max_steps, trials):
    current_trial = 0
    steps = [i for i in range(max_steps+1)]
    #steps_list = []
    #steps_list.extend(steps*trials)
    position_list = []
    while current_trial < trials:
        position = [0]
        current_step = 1
        while current_step < max_steps + 1:
            if np.random.random() >= 0.5:
                position.append(position[current_step - 1] + 1)
            else:
                position.append(position[current_step - 1] - 1)
            current_step += 1
        position_list.extend([position])
        current_trial += 1
        
    return steps, position_list

steps, position_list = walk_int_steps(16, 1000)

k = 0
while k<len(position_list):
    plt.plot(steps, position_list[k], linewidth=0.5, alpha=0.1)
    k += 1

plt.show()
```

```python
#just an exercise to basically replicate the binomial distribution, I think.
def walk_cont_steps(max_steps, trials):
    current_trial = 0
    steps = [i for i in range(max_steps+1)]
    #steps_list = []
    #steps_list.extend(steps*trials)
    position_list = []
    while current_trial < trials:
        position = [0]
        current_step = 1
        while current_step < max_steps + 1:
            x = np.random.uniform(-1,1)
            position.append(position[current_step - 1] + x)
            current_step += 1
        position_list.extend([position])
        current_trial += 1
        
    return steps, position_list

steps, position_list = walk_cont_steps(16, 100)

fig_1 = plt.figure(2, figsize=(20, 8))

chart_1 = fig_1.add_subplot(211)
k = 0
while k<len(position_list):
    chart_1.plot(steps, position_list[k], linewidth=0.5, alpha=0.1, color='b')
    k += 1
chart_1.set_ylabel('position')
chart_1.set_xlabel('step number')
chart_1.set_title("1000 random walks")

plt.show()
```

```python
#R code 4.16
#I'm just going to copy this from the python git I've been following. Even in the text, Richard McElreath says
#the R version of this code isn't worth delving into because it requires some tricks that don't provide conceptual
#understanding. We're just developing a 2D grid approximation this one time to get some data to compare to 
#the data our model produces later on.
post = np.mgrid[140:160:0.1, 4:9:0.1].reshape(2,-1).T # this looks like it returns a 2D array of means and stdevs

likelihood = [sum(stats.norm.logpdf(d2.height, loc=post[:,0][i], scale=post[:,1][i])) for i in range(len(post))]

post_prod = (likelihood + 
             stats.norm.logpdf(post[:,0], loc=178, scale=20) + 
             stats.uniform.logpdf(post[:,1], loc=0, scale=50))
post_prob = np.exp(post_prod - max(post_prod))
```

```python
#post is a 2D array
#post[:] is equivalent to post
#post[2:5, 1] returns the data from rows 2-5 in column 1 (not 0)
#post[:,0] returns all the values in column 0
#post[:,0].min() returns the minimum value in column 0
```

```python
#read notes above when you forget the syntax. The two lines below are just short ways of setting the linspace
#relative to the min and max values for mean and stdev in post
xi = np.linspace(post[:,0].min(), post[:,0].max(), 100) 
yi = np.linspace(post[:,1].min(), post[:,1].max(), 100)
#griddata from scipy interpolates unstructured data. We only know a few points, but we need lots more
zi = griddata((post[:,0], post[:,1]), #all of the mus, all of the sigmas
              post_prob,  # posterior probability from above 
              (xi[None,:], yi[:,None])) # this has something to do with the width of the interpolated data

plt.contour(xi, yi, zi);
```

```python
with pm.Model() as m4_5:
    alpha = pm.Normal('alpha', mu=178, sd=20) 
    beta = pm.Lognormal('beta', mu=0, sd=1) 
    mu = pm.Deterministic('mu', alpha + beta*(d2.weight - d2.weight.mean())) # by including mu like this, it shows up in the posterior sample
    sigma = pm.Uniform('sigma', lower=0, upper=50)
    height = pm.Normal('height', mu=mu, sd=sigma, observed=d2.height)
    trace_4_5 = pm.sample(1000, tune=1000) 
```

```python
weight_seq = np.arange(25, 71) # returns an array of integers 25-70 corresponding to weights
# Given that we have a lot of samples we can use less of them for plotting (or we can use all!)
trace_4_5_thinned = trace_4_5[::10]
mu_pred = np.zeros((len(weight_seq), len(trace_4_5_thinned)*trace_4_5.nchains))

#the above line returns an array with 46 rows and 400 columns. The 46 rows are because we're looking at the 46 weight
#values we pass in from weight_seq. The 400 columns are because trace_4_5_thinned has length 100 and there are 4 chains
#in trace_4_5

for i, w in enumerate(weight_seq):
    mu_pred[i] = trace_4_5_thinned['alpha'] + trace_4_5_thinned['beta'] * (w - d2.weight.mean())

#each row in mu_pred represents one real weight. Each column represents a sample from the posterior distribution.
#Each element represents the average height, factoring in uncertainty for alpha and beta, for the given weight

pd.DataFrame(mu_pred)
```

```python
#R code 4.65
#We're going to start by normalizing the weights and the squares of the weights by adding columns to d. Essentially, turn
#them into z-scores.

d['weight_norm'] = (d.weight - d.weight.mean())/d.weight.std()
d['weight_norm_sq'] = d.weight_norm**2

with pm.Model() as m4_6:
    alpha = pm.Normal('alpha', mu=178, sd=20) 
    beta_1 = pm.Lognormal('beta_1', mu=0, sd=1) 
    beta_2 = pm.Normal('beta_2', mu=0, sd=1)
    mu = pm.Deterministic('mu', alpha + beta_1*d.weight_norm + beta_2*d.weight_norm_sq) # by including mu like this, it shows up in the posterior sample
    sigma = pm.Uniform('sigma', lower=0, upper=50)
    height = pm.Normal('height', mu=mu, sd=sigma, observed=d.height)
    trace_4_6 = pm.sample(1000, tune=1000) 
```

```python
#part (a)
with pm.Model() as m_4H2:
    alpha = pm.Normal('alpha', mu=70, sd=30) 
    beta = pm.Lognormal('beta', mu=0, sd=1) 
    mu = pm.Deterministic('mu', alpha + beta*(d3.weight - d3.weight.mean())) # by including mu like this, it shows up in the posterior sample
    sigma = pm.Uniform('sigma', lower=0, upper=50)
    height = pm.Normal('height', mu=mu, sd=sigma, observed=d3.height)
    trace_4H2 = pm.sample(1000, tune=1000) 
```

```python
#4H3
with pm.Model() as m_4H3:
    alpha = pm.Normal('alpha', mu=70, sd=30) 
    beta = pm.Lognormal('beta', mu=0, sd=1) 
    mu = pm.Deterministic('mu', alpha + beta*np.log(d.weight )) # by including mu like this, it shows up in the posterior sample
    sigma = pm.Uniform('sigma', lower=0, upper=50)
    height = pm.Normal('height', mu=mu, sd=sigma, observed=d.height)
    trace_4H3 = pm.sample(1000, tune=1000) 
```

```python
varnames = ['alpha', 'beta','sigma']
pm.summary(trace_4H3, varnames, alpha = 0.11).round(2)
```

```python
a_mean_all = trace_4H3['alpha'].mean()
b_mean_all = trace_4H3['beta'].mean()

weight_seq_all = np.arange(4, 70) # returns an array of integers 4-70 corresponding to weights
# Given that we have a lot of samples we can use less of them for plotting (or we can use all!)
trace_4H3_thinned = trace_4H3[::20]
mu_pred_all = np.zeros((len(weight_seq_all), len(trace_4H3_thinned)*trace_4H3.nchains))

for i, w in enumerate(weight_seq_all):
    mu_pred_all[i] = trace_4H3_thinned['alpha'] + trace_4H3_thinned['beta'] * np.log(w)

#each row in mu_pred represents one real weight. Each column represents a sample from the posterior distribution.
#Each element represents the average height, factoring in uncertainty for alpha and beta, for the given weight

pd.DataFrame(mu_pred_all)

height_pred_all = pm.sample_posterior_predictive(trace_4H3, 200, m_4H3)

mu_mean_all = mu_pred_all.mean(1)  #calculate the mean along the first axis of mu_pred, returns an array
mu_all_hpd = pm.hpd(mu_pred_all.T, alpha=.03)

heights_all_hpd = pm.hpd(height_pred_all['height'], alpha=.11)

#we have to order the weights in d2.weight just to make the shading make sense.
idx = np.argsort(d.weight.values)
d_weight_ord = d.weight.values[idx]
heights_all_hpd = heights_all_hpd[idx]


figure_9 = plt.figure(figsize=(8,6))
chart_14 = figure_9.add_subplot(111)
chart_14.plot(d.weight, d.height, '.')

chart_14.fill_between(d_weight_ord, heights_all_hpd[:,0], heights_all_hpd[:,1], color='C2', alpha=0.25)
chart_14.set_xlabel('weight')
chart_14.set_ylabel('height');
# chart_13.set_title('Data (scatter) with posterior average values of alpha and beta determining the line')
chart_14.plot(weight_seq_all, mu_mean_all)
chart_14.fill_between(weight_seq_all, mu_all_hpd[:,0], mu_all_hpd[:,1],color='r',alpha=0.25);

#I'd say that's pretty damn good.
```
